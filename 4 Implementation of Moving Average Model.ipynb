## 4 Implementation of Moving Average Model 
### (14/12/2024)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dates=pd.date_range(start='2024-01-01',end='2024-01-31',freq='d').tolist()
num=[43,31,1,10,20,24,26,27,34,35,36,37,31,21,20,19,18,20,24,28,29,20,32,31,34,35,36,30,32,30,23]
```

```python
len(num)
```

```python
df=pd.DataFrame({'Dates':dates,'Price':num})
```

```python
df.head(10)
```

```python
plt.plot(df['Dates'],df['Price'])
plt.xlabel('Dates')
```

```python
df.plot.line(x='Dates',y='Price')
```

## 3 Year Moving average

```python
df1=pd.DataFrame({'Dates':dates,'Price':num})
df1['Moving Average']=df1['Price'].rolling(window=3).mean()
```

```python
df1
```

```python
plt.plot(df1['Dates'],df1['Price'],label="Price")
plt.plot(df1['Dates'],df1['Moving Average'],label="Moving average")
plt.xlabel('Dates')
plt.legend()
```

```python
df1.plot.line(x="Dates",y=["Price","Moving Average"])
```

```python
df1['Rolling Average Center']=df1['Price'].rolling(window=3,center=True).mean()
```

```python
df1
```

```python
df1.plot.line(x="Dates",y=["Price","Rolling Average Center"])
```
