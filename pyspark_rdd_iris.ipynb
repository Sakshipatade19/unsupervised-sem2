```python
from pyspark.context import SparkContext
sc= SparkContext("local","iris")
```

```python
iris1=sc.textFile("iris.csv",2)
iris1.collect()
```

```python
partitioned_data=iris1.glom().collect()
for i, partition in enumerate(partitioned_data):
  print(f"Partiton-{i+1} has {len(partition)} records")
```

```python
iris2=sc.parallelize(iris1.collect(),8)
iris2.getNumPartitions()
```

```python
for i in iris2.collect():
  print(i)
```

```python
header=iris2.first()
str_to_igr=header
```

```python
iris2=iris2.filter(lambda x:x!=str_to_igr)
iris2.collect()
```

```python
iris2.count()
```

```python
iris3=iris2.map(lambda x:x*2)
iris3.collect()
```

```python
iris4=iris2.flatMap(lambda x:x.split(","))
iris4.collect()
```

```python
partitons = iris4.glom().collect()
print("Total Partitions : ",len(partitons))
```

```python
partitons[0]
```

```python
iris5 = iris4.map(lambda x:x*2)
```

```python
iris4.take(5)
```

```python
iris6 =iris1.flatMap(lambda x:x.split(",")).filter(lambda x : x=='"virginica"').map(lambda x: (x,1))

```

```python
partitions = iris6.glom().collect()
partitions
```

```python
species =['"virginica"','"setosa"','"versicolor"']
species1 =iris1.flatMap(lambda x:x.split(",")).filter(lambda x : x in species)
species1.collect()
```

```python
species2 = species1.groupBy(lambda x:x).mapValues(lambda x:len(x))
species2.collect()
```
