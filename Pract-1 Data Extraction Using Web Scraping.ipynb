# Data Extraction Using Web Scraping 20/1/24

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests as rs
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
src = rs.get(url)
print("Status code:", src.status_code)
```

```python
soup = BeautifulSoup(src.content, features = "lxml") #parsing tg
soup
```

```python
soup.prettify()
```

```python
soup.body()
```

```python
t1 = soup.find('table')
t1
```

```python
cols = []
t_header = t1.find('thead')
t_header
```

```python
for tr in t_header.find_all('tr'):
    cols.append(tr.text)
cols
```

```python
'''cols = []
for i in t1.find_all('thead'):
    cols.append(i.text)
cols'''
```

```python
t_body = t1.find('tbody')
t_rows = t_body.find_all('tr')
d1 = []
```

```python
for rows in t_rows:
    data = []
    for row_data in rows.find_all('td'):
        data.append(row_data.text)
    d1.append(data)
d1
```

```python
df = pd.DataFrame(d1)
df.head()
```

```python
df.to_csv('output.csv')
```
