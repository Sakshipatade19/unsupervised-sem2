```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as exp
```

```python
df = pd.read_csv('/content/drive/MyDrive/MSC DSAI/sem-2/Unsupervised  Learning/pract-1/mnist_train.csv')
df.head()
```

```python
df = pd.read_csv('/content/drive/MyDrive/MSC DSAI/sem-2/Unsupervised  Learning/pract-1/mnist_test.csv')
df.head()
```

```python
x = df.drop('label', axis = 1)
x.head()
```

```python
y = df['label']
y.head()
```

PCA

```python
pca = PCA(n_components = 2)
X_two = pca.fit_transform(x)
print(X_two)
```

```python
pca = PCA(n_components = 3)
X_three = pca.fit_transform(x)
print(X_three)
```

```python
plt.figure(figsize = [8,6])
plt.title("2d Scatter plot")
plt.scatter(X_two[:,0], X_two[:, 1], c=y)
plt.show()
```

```python
fig = plt.figure(figsize = [8,6])
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_three[:, 0], X_three[:, 1], X_three[:, 2], c=y, cmap='viridis', s=60)
ax.set_title("3d scatter plot")
ax.set_xlabel("1st Principle Component")
ax.set_ylabel("2nd Principle Component")
ax.set_zlabel("3rd Principle Component")
plt.show()
```

```python
fig = exp.scatter_3d(x = X_three[:, 0], y = X_three[:, 1], z = X_three[:, 2], color=y)
fig.show()
```
