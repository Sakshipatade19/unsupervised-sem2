```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
```

```python
df=pd.read_csv('C:/Users/uday8/Documents/Msc. DSAI/SEM 2/TSA/Data/AirPassengers.csv',
               parse_dates=['Month'],index_col='Month')
df.head()
```

```python
plt.figure(figsize=(7,4))
plt.plot(df.index,df['#Passengers'],label='Original time series',color='green')
plt.legend(loc='best')
plt.title('Air passenger dataset')
plt.xlabel('year')
plt.ylabel('Number of passengers')
plt.show()
```

```python
# perform STL decomposition
stl = STL(df['#Passengers'], seasonal=13, robust=True)
res = stl.fit()
fig,(ax1,ax2,ax3)=plt.subplots(3,1,figsize=(7,4))
ax1.plot(df.index,res.trend, label='Trend',color='red')
ax1.set_title('Trend Component')
ax2.plot(df.index,res.seasonal, label='Seasonal',color='blue')
ax2.set_title('Seasonal Component')
ax3.plot(df.index,res.resid, label='Residual',color='green')
ax3.set_title('Residual Component')
plt.tight_layout()
plt.show()
```
